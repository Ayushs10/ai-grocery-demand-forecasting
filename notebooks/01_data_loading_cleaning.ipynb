## 5. Check Missing Values

## 6. Handle Missing Values

`days_since_prior_order` is null for a customer's first order, so we replace it with 0.

```python
orders['days_since_prior_order'] = orders['days_since_prior_order'].fillna(0)
```

```python
orders.isnull().sum()
```

## 7. Check for Duplicates

```python
print("Duplicate rows in orders:", orders.duplicated().sum())
print("Duplicate rows in order_products:", order_products.duplicated().sum())
print("Duplicate rows in products:", products.duplicated().sum())
print("Duplicate rows in aisles:", aisles.duplicated().sum())
print("Duplicate rows in departments:", departments.duplicated().sum())
```

## 8. Inspect Data Types

```python
print(orders.dtypes)
print()
print(order_products.dtypes)
print()
print(products.dtypes)
print()
print(aisles.dtypes)
print()
print(departments.dtypes)
```

## 9. Optional Memory Optimization

```python
orders['order_dow'] = orders['order_dow'].astype('int8')
orders['order_hour_of_day'] = orders['order_hour_of_day'].astype('int8')
orders['order_number'] = orders['order_number'].astype('int16')

order_products['add_to_cart_order'] = order_products['add_to_cart_order'].astype('int16')
order_products['reordered'] = order_products['reordered'].astype('int8')

products['aisle_id'] = products['aisle_id'].astype('int16')
products['department_id'] = products['department_id'].astype('int8')
```

## 10. Validate Important Columns

```python
print("Unique order_dow values:", sorted(orders['order_dow'].unique()))
print("Unique order_hour_of_day values:", sorted(orders['order_hour_of_day'].unique()))
```

## 11. Merge Orders with Order Products

This adds order time information to each product-level row.

```python
orders_products = order_products.merge(
    orders[['order_id', 'order_dow', 'order_hour_of_day', 'days_since_prior_order']],
    on='order_id',
    how='left'
)
```

```python
orders_products.head()
```

```python
orders_products.shape
```

## 12. Merge Product Information

```python
orders_products = orders_products.merge(
    products[['product_id', 'product_name', 'aisle_id', 'department_id']],
    on='product_id',
    how='left'
)
```

## 13. Merge Aisle Information

```python
orders_products = orders_products.merge(
    aisles,
    on='aisle_id',
    how='left'
)
```

## 14. Merge Department Information

```python
orders_products = orders_products.merge(
    departments,
    on='department_id',
    how='left'
)
```

```python
orders_products.head()
```

```python
orders_products.shape
```

## 15. Rename Columns for Clarity

```python
orders_products = orders_products.rename(columns={
    'order_dow': 'day_of_week',
    'order_hour_of_day': 'hour',
    'department': 'dept_name',
    'aisle': 'aisle_name',
    'reordered': 'is_reordered',
    'add_to_cart_order': 'cart_position'
})
```
